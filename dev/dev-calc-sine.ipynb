```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# User specified number of data points
n_points = 50

# Generate x and y vectors
x = np.around(np.arange(0.0, 5.0, 5.0/n_points),decimals=2)
y = np.sin(2*np.pi*x)

# Plot results
plt.plot(x, y)
plt.title('Since function')
plt.show()
```

```python
def calc_sine(n_points=10):
    x = np.around(np.arange(0.0, 5.0, 5.0/n_points),decimals=2)
    y = np.sin(2*np.pi*x)
    return x, y
```

```python
# Test function
x,y = calc_sine(100)

# Plot results
plt.plot(x, y)
plt.title('Since function')
plt.show()
```
